```python
import re
train_tuple = open("tuples_train_posi.txt")
neg_train_tuple = open("tuples_train_neg.txt")
train_top_tmp = open("top_ind_train.txt") 
train_top_jpg = open("../img_list/img_list_top.txt")
train_top_tmp_content = train_top_tmp.readlines()
print(len(train_top_tmp_content))
train_top_jpg_content = train_top_jpg.readlines()
print(len(train_top_jpg_content))
train_top_list = open('../../../train_top.txt','w')
```

```python
top_i1, top_i2 = 0, 0
count = 0
for line in train_tuple:
    count = count + 1
    top_i1 = re.findall(r'\d+',line)[1]
    top_i2 = train_top_tmp_content[int(top_i1)]
    top_path = train_top_jpg_content[int(top_i2)].strip('\r\n') + ' 1\r\n'
    print(top_path)
    train_top_list.write(top_path)
for neg_line in neg_train_tuple:
    count = count - 1
    if count < 0:
        break
    top_i1 = re.findall(r'\d+',line)[1]
    top_i2 = train_top_tmp_content[int(top_i1)]
    top_path = train_top_jpg_content[int(top_i2)].strip('\r\n') + ' 0\r\n'
    print(top_path)
    train_top_list.write(top_path)
train_tuple.close()
neg_train_tuple.close()
train_top_tmp.close()
train_top_jpg.close()
train_top_list.close()
print(count)
```

```python
#new
import re
train_tuple = open("tuples_train_shuffled.txt")
train_top_tmp = open("top_ind_train.txt") 
train_top_jpg = open("../img_list/img_list_top.txt")
train_top_tmp_content = train_top_tmp.readlines()
print(len(train_top_tmp_content))
train_top_jpg_content = train_top_jpg.readlines()
print(len(train_top_jpg_content))
train_top_list = open('../../../train_top.txt','w')
```

```python
top_i1, top_i2 = 0, 0
count = 0
for line in train_tuple:
    count = count + 1
    top_i1 = re.findall(r'\d+',line)[1]
    top_i2 = train_top_tmp_content[int(top_i1)]
    if (re.findall(r'\d+',line)[4]) == '1':
        top_path = train_top_jpg_content[int(top_i2)].strip('\r\n') + ' 1\r\n'
    elif re.findall(r'\d+',line)[4] == '0':
        top_path = train_top_jpg_content[int(top_i2)].strip('\r\n') + ' 0\r\n'
    train_top_list.write(top_path)

train_tuple.close()
train_top_tmp.close()
train_top_jpg.close()
train_top_list.close()
print(count)
```
